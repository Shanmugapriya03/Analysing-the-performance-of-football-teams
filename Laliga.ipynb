```python
import pandas as pd
import numpy as np
```

## Read the dataset and replace dashes with 0

```python
Laliga = pd.read_csv('Laliga.csv',header=1)
Laliga.head()
```

```python
Laliga.shape
```

```python
Laliga.replace('-',0,inplace=True)
```

## Print all the teams which have started playing between 1930-1980.

### Splitting the 'Debut' column with '-' to get start and end year

```python
Laliga['Debut_Split'] = Laliga['Debut'].str.split('-')
Laliga.head()
```

### Splitting the list values (Start and end year) in to 2 columns in a new data frame 'Laliga_debut'

```python
Laliga_debut = pd.DataFrame(Laliga['Debut_Split'].values.tolist(), columns=['Debut_Start','Debut_End'])
Laliga_debut.head()
```

### Cleaning the data
#### Replacing the None values with the start year
#### Replacing the end year values in the same format of start year ( prefix of start year is used)

```python
len = Laliga.shape[0]
for i in range(len):
    if (Laliga_debut['Debut_End'][i] != None):
        Laliga_debut['Debut_End'][i] = Laliga_debut['Debut_Start'][i][:2] + Laliga_debut['Debut_End'][i]
    else :
        Laliga_debut['Debut_End'][i] = Laliga_debut['Debut_Start'][i]
```

```python
Laliga_debut.head()
```

### Merging the 2 DataFrame 'Laliga' & 'Laliga_debut'

```python
Laliga = Laliga.join(Laliga_debut)
```

### Drop the column 'Debut_Split'

```python
Laliga.drop(['Debut_Split'],axis=1,inplace=True)
```

```python
Laliga.dtypes
```

```python
Laliga['Debut_Start'] = Laliga.Debut_Start.astype(int)
Laliga['Debut_End'] = Laliga.Debut_End.astype(int)
```

```python
Laliga.dtypes
```

```python
Laliga[(Laliga['Debut_Start'] > 1930) & (Laliga['Debut_End'] <1980)]
```

### Drop the columns 'Debut_Start' and 'Debut_End'

```python
Laliga.drop(['Debut_Start','Debut_End'],axis=1,inplace=True)
```

```python
Laliga.head()
```

## Write a function with name â€œGoal_diff_countâ€ which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference

#### Converting the datatype of 'GoalsFor' and 'GoalsAgainst' column to int

```python
Laliga['GoalsFor'] = Laliga.GoalsFor.astype(int)
Laliga['GoalsAgainst'] = Laliga.GoalsAgainst.astype(int)
```

```python
def Goal_diff_count():
    Laliga['Goal_diff_count'] = Laliga.loc[:,'GoalsFor'] - Laliga.loc[:,'GoalsAgainst']
    print('Maximun goal difference :', Laliga['Goal_diff_count'].max())
    print('Minimun goal difference :', Laliga['Goal_diff_count'].min())
Goal_diff_count()
```

### Drop the column 'Goal_diff_count'

```python
Laliga.drop(['Goal_diff_count'],axis=1,inplace=True)
```

## Create a new column with name â€œWinning Percentâ€ and append it to the data set

```python
Laliga['GamesWon'] = Laliga.GamesWon.astype(int)
Laliga['GamesPlayed'] = Laliga.GoalsAgainst.astype(int)
```

### Percentage of Winning = (GamesWon / GamesPlayed)*100

```python
Laliga['WinningPercent'] = (Laliga.loc[:,'GamesWon'] / Laliga.loc[:,'GamesPlayed'])*100
Laliga['WinningPercent'].head()
```

### If there are any numerical error, replace it with 0%

```python
Laliga['WinningPercent'].fillna(0,inplace=True)
```

```python
Laliga.tail()
```

### Print the top 5 teams which has the highest Winning percentage

```python
Laliga.sort_values(by=['WinningPercent'],ascending=False,inplace=True)
```

### Group teams based on their â€œBest positionâ€ and print the sum of their points for all positions
##### Eg: Best_Position            Points
##### 1                              25000
##### 2                              7000   

```python
LaligaGroup = Laliga[['BestPosition','Points']]
```

```python
LaligaGroup.head()
```

```python
GroupByBestPosition = LaligaGroup.groupby('BestPosition')
```

```python
print("Best_Position","\tPoints")
for BestPosition,data in GroupByBestPosition:
    totalPoints = 0
    for i in data['Points']:
        totalPoints += i
    print(BestPosition,"\t\t",totalPoints)
```
